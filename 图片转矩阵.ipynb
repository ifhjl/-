```python
import pandas as pd
import numpy as np
from PIL import Image,ImageFilter
from matplotlib import pyplot as plt
from itertools import combinations
import os
```

```python
def image_to_matrix(image_path):
    img = Image.open(image_path)
    img = img.convert('L')
    img_matrix = np.array(img)
    img_matrix = np.delete(img_matrix, np.where(np.all(img_matrix == 255, axis=1)), axis=0)
    img_matrix = np.delete(img_matrix, np.where(np.all(img_matrix == 255, axis=0)), axis=1)
    img_matrix = Image.fromarray(img_matrix)
    img_matrix = img_matrix.resize((28,28))
    img_matrix = np.array(img_matrix).reshape(1,-1)
    return img_matrix[0]
```

```python
path = ["我的手写字母/"+i for i in os.listdir("我的手写字母")]
path
```

```python
data_ = [image_to_matrix(i) for i in path]
```

```python
d = pd.DataFrame(data_)
#将d中为负数的值加256
d[d<0] = d[d<0] + 256
```
